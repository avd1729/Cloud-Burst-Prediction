```python
data['sunrise'] = pd.to_datetime(data['sunrise']).dt.strftime('%H:%M:%S')
data['sunrise'] = pd.to_datetime(data['sunrise'], format='%H:%M:%S')

print(data['sunrise'].head())
```

```python
data['sunset'] = pd.to_datetime(data['sunset']).dt.strftime('%H:%M:%S')
data['sunset'] = pd.to_datetime(data['sunset'], format='%H:%M:%S')

print(data['sunset'].head())
```

```python
# Replace 'No moonrise' with NaN
data['moonrise'] = data['moonrise'].replace('No moonrise', np.nan)

# Convert valid entries to datetime
data['moonrise'] = pd.to_datetime(data['moonrise'], format='%I:%M %p', errors='coerce')

print(data['moonrise'].head())
```

```python
# Replace 'No moonrise' with NaN
data['moonset'] = data['moonset'].replace('No moonrise', np.nan)

# Convert valid entries to datetime
data['moonset'] = pd.to_datetime(data['moonset'], format='%I:%M %p', errors='coerce')

print(data['moonset'].head())
```
